```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/drive/MyDrive/GreenTrail Store Data.xlsx")
print("Data loaded successfully!")
```

```python
# prompt: draw a graph for average percentage increase in sales from before to during promotions . xlabel - distinct(type of promotion), ylabel = overall avg percentage increase in sales

# Group by promotion type and calculate the average percentage increase in sales
promotion_sales = df.groupby('Type of Promotion')['Percentage Increase in Sales'].mean().reset_index()

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Promotion Type', y='Percentage Increase in Sales', data=promotion_sales)
plt.xlabel('Distinct Promotion Type')
plt.ylabel('Overall Average Percentage Increase in Sales')
plt.title('Average Percentage Increase in Sales per Promotion Type')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
urban_stores = df[df['Store Type'] == 'Urban']['Store ID'].count()
suburban_stores = df[df['Store Type'] == 'Suburban']['Store ID'].count()

print(f"Number of urban stores: {urban_stores}")
print(f"Number of suburban stores: {suburban_stores}")
```
